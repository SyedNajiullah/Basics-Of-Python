#Importing

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

# Matplot

Basic figure simple line ploting using matplot

```python
x_values = [1, 2, 3, 4, 5]
y_values = [1, 4, 3, 2, 5]
plt.plot(x_values, y_values)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Example Graph")
plt.show()
```

Multiple plotting

```python
x_values = np.linspace(1, 10, 5)
y_values1 = np.sin(x_values)
y_values2 = np.cos(y_values)

plt.plot(x_values, y_values1, label="sin(x)")
plt.plot(x_values, y_values2, label="cos(x)")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.title("Sin Cosine curve")
plt.show()
```

Scatter plot

```python
x_values = np.linspace(1, 15, 10)
y_values = np.linspace(1, 45, 10)

plt.scatter(x_values, y_values)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Scatter Plot")
plt.show()
```

Colors line width, type

```python
x_values = np.linspace(0, 10, 20)
y_values1 = np.sin(x_values)
y_values2 = np.cos(x_values)
y_values3 = np.tan(x_values / 2)

plt.plot(x_values, y_values1, label="sin(x)", color="green", linewidth=5, linestyle=":", alpha = 0.7) # Dotted line, alpha = transparence
plt.plot(x_values, y_values2, label="cos(x)", color="red", linewidth=5, linestyle="--", alpha=1) # Dashed line
plt.plot(x_values, y_values3, label="tan(x/2)", color="blue", linewidth=5, linestyle="-", alpha=0.5) # straignt line

plt.legend()
plt.xlim(0, 10)
plt.ylim(-2, 2)

plt.title("Sin Cos Tan")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.show()
```

Color scatter plot

```python
x_values = np.linspace(0, 10, 100)
y_values = np.sin(x_values)
colors_id = np.linspace(0, 1, len(x_values))

plt.scatter(x_values, y_values, c=colors_id, cmap="viridis")
plt.colorbar(label="Colorbar")

plt.show()
```

```python
figure, axes = plt.subplots(2, 2, figsize=(8, 8))

x_values = np.linspace(0,10, 10)

axes[0, 0].plot(x_values, np.sin(x_values), "r-")
axes[0, 0].set_xlabel("x-axis")
axes[0, 0].set_ylabel("y-axis")
axes[0, 0].set_title("six(x)")

axes[0, 1].plot(x_values, np.cos(x_values), "g-")
axes[0, 1].set_xlabel("x-axis")
axes[0, 1].set_ylabel("y-axis")
axes[0, 1].set_title("cos(x)")

axes[1, 0].plot(x_values, np.tan(x_values), "b-")
axes[1, 0].set_xlabel("x-axis")
axes[1, 0].set_ylabel("y-axis")
axes[1, 0].set_title("tan(x)")

axes[1, 1].plot(x_values, np.tanh(x_values), "y-")
axes[1, 1].set_xlabel("x-axis")
axes[1, 1].set_ylabel("y-axis")
axes[1, 1].set_title("tanh(x)")

figure.tight_layout()
figure.show()
```

Bar graph, Horizontal Bar graph

```python
x_categories = ["A", "B", "C", "D"]
y_values = [25, 50, 10, 15]

plt.bar(x_categories, y_values, color=["Red", "Green", "Blue", "Yellow"])
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Graph")
plt.show()

plt.barh(x_categories, y_values, color=["Red", "Green", "Blue", "Yellow"])
plt.xlabel("Values")
plt.ylabel("Categories")
plt.title("Bar Graph")
plt.show()
```

Histograms, Group histogram

```python
values = [70, 10, 34]
print(values)
plt.hist(values, color="green", alpha = 0.7)
plt.xlabel("x_axis")
plt.ylabel("y_axis")
plt.title("Histogram")
plt.show()


values1 = [1, 5, 3, 2, 1, 2]
values2 = [3, 0, 5, 9, 10, 2]

plt.hist(values1, color="blue", alpha=0.5)
plt.hist(values2, color="red", alpha=0.5)
plt.xlabel("x_axis")
plt.ylabel("y_axis")
plt.title("Grouped Histogram")
plt.show()
```

Pie graph

```python
pie_values = [10, 25, 10, 15, 40]
pie_categroies = ["A", "B", "C", "D", "E"]
colors = ["Red", "Green", "Blue", "Orange", "Black"]

plt.pie(pie_values, labels=pie_categroies, colors=colors)
plt.title("Pie Graph")
plt.legend()
plt.show()
```

Heatmaps

```python
data = np.array([
    [1, 2, 3], [4, 5, 6]
])

plt.imshow(data, cmap="hot")
plt.colorbar()
plt.title("Heat map")
plt.show()
```

```python
penguins_dataframe = sns.load_dataset("penguins")
penguins_dataframe.head()
```

```python
# Scatter Plot on flipper_length_mm vs body_mass_g grouped by island
sns.scatterplot(data=penguins_dataframe, x="flipper_length_mm", y="body_mass_g", hue = "island")
```

```python
# Scatter Plot on flipper_length_mm vs body_mass_g grouped by gender
sns.set_style("ticks") # allowed styles are white, dark, whitegrid, darkgrid, ticks
sns.set_context("notebook") # allowed context are paper, notebook, talk, poster
sns.scatterplot(data=penguins_dataframe, x="flipper_length_mm", y="body_mass_g", hue = "sex")
sns.despine() # removing top and left lines
```

```python
sns.scatterplot(data=penguins_dataframe, x="species", y="body_mass_g", hue = "island", style = "sex", alpha=0.7)
```
