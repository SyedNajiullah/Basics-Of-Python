#Importing

```python
import pandas as pd
import numpy as np
```

Converting numpy array into pandas dataframe

```python
arr = np.array([
    [' ',"Book", "Type"],
     ['0',"Lord Of The Ring", "Good"],
      ['1',"Alchemist", "Normal"],
       ['2',"Think and grow rich", "Normal"],
        ['3', 'Dune', 'Science'],
         ['4', 'Educated', 'Memoir'],
          ['5', 'Hobbit', 'Fantasy']
    ])
arr
```

```python
data_frame = pd.DataFrame(data=arr[1:, 1:], columns=arr[0, 1:3], index=arr[1:7, 0])
data_frame
```

```python
print(data_frame.head())
print()
print(data_frame.tail())
```

Renaming a column


```python
data_frame.rename(columns={"Type":"Genre", "Book": "Title"}, inplace=True)
data_frame
```

Adding a new column

```python
data_frame["Author"] = np.nan
data_frame
```

Deleting a column

```python
del data_frame["Author"]
data_frame
```

Changing names of the data

```python
data_frame["Genre"] = data_frame["Genre"].replace({"Science": "Science Fiction"})
data_frame
```

Selecting specific data

```python
df = data_frame[data_frame["Genre"] == "Normal"]
df
```

Diference between loc and iloc

loc is used to select rows but with the name of the row
iloc is used to select rw but with the name of index of the row

```python
print(data_frame)
print("-"*5)
print("Loc")
print("-"*5)
# iloc
print(data_frame.loc["2"])
print()
print(data_frame.loc["1": "4"]) # From 1 till 4
# loc
print("-"*5)
print("ilLoc")
print("-"*5)
print(data_frame.iloc[2])
print()
print(data_frame.iloc[1: 4]) # From 1 till 3
```

Reading auto csv file

```python
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
auto_data_frame = pd.read_csv("/content/auto.csv", names=headers)
auto_data_frame.head()
```

replacing "?" with np.nan

```python
auto_data_frame.replace("?", np.nan, inplace=True)
auto_data_frame.head()
```

counting missing data

```python
missing_data = auto_data_frame.isnull()
print(missing_data)
missing_data.sum()
```

Replacing nan values with the None

```python
auto_data_frame["normalized-losses"].replace(np.nan, None , inplace = True)
print(auto_data_frame)
```

Drop any null rows in all over the dataset

```python
auto_data_frame = auto_data_frame.dropna()
auto_data_frame
```

Since some rows are removed so now the we need to reset in indexes of the rows

```python
auto_data_frame = auto_data_frame.reset_index(drop=True)
auto_data_frame
```

```python
df = pd.DataFrame({"names": ["Harry", "Safi", "Zeerak"], "marks":[45, 95, 125]})
df
```

```python
def get_percentage(marks):
  return (marks / 130) * 100

df["Percentage"] = df["marks"].apply(get_percentage)
df
```

```python
df["marks"] = df["marks"] + 5
df
```
