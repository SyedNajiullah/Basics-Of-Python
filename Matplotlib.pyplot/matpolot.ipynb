```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([1,2,3])
y = np.array([4,5,6])
plt.plot(x,y)
plt.xlabel("X_AXIS")
plt.ylabel("Y_AXIS")
plt.title("SOME_THING")
plt.show()
```

```python
y1Points = np.array([4,5,6])
y2Points = np.array([1,2,9])
plt.plot(y1Points)
plt.plot(y2Points)
```

```python
xPoints = np.array(['A', 'B', 'C'])

yPoints = np.array([121,22,122])
plt.bar(xPoints,yPoints)
```

```python
arr = np.array([1,234,53])
mean = np.mean(arr)
standard = np.std(arr)
x = np.random.normal(mean, standard, 200)
plt.hist(x)
plt.show()
```
